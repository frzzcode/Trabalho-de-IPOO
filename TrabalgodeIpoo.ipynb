```python
import math

def vet_reta_reta():
  #Criando os vetores do codigo
  vetor_u = []
  vetor_v = []
  vetprod = []

  #Receber os dados dos vetores que o usuario passar
  #Primeiro Vetor
  for x in range(3):
    valor = float(input("Digite o valor de x,y,z do Primeiro vetor: "))
    vetor_u.append(valor)
  #Segundo Vetor
  for x in range(3):
    valor = float(input("Digite o valor de x,y,z do Segundo vetor: "))
    vetor_v.append(valor)

  #Mostrar os vetores Recebidos
  print("O primero vetor eh", vetor_u)
  print("O Segundo vetor eh", vetor_v)

  #Calcular o Produto escalar dos vetores
  for i in range(3):
    prod_escalar = vetor_u[i] * vetor_v[i]
    vetprod.append(prod_escalar)
  somas = sum(vetprod)

  #Mostra o Produto Escalar
  print("O produto escalar dos vetores eh", somas)

  #Modulo eh a raiz quadrada da soma dos indices ao quadrado, ent tenho q pegar e fazer os indices ao quadrado e soma-los, dps fazer a raiz deles

  modulo_vet1 = math.sqrt(vetor_u[0]**2 + vetor_u[1]**2 + vetor_u[2]**2)
  modulo_vet2 = math.sqrt(vetor_v[0]**2 + vetor_v[1]**2 + vetor_v[2]**2)

  #Resultado na formula
  cos = somas / (modulo_vet1 * modulo_vet2)

  #Resultado Final Utilizando arccos, Utiliazdo a funlção degrees para tranformar em graus.
  ang = math.degrees(math.acos(cos))

  #Mostrar o Modulo do Primeiro vetor
  print("O modulo do Primeiro vetor eh", modulo_vet1)
  #Mostrar o Modulo do Segundo vetor
  print('O modulo do Segundo vetor eh', modulo_vet2)
  #Resultado final, em Angulo
  print("O Angulo entre os vetores eh", ang)

while True:

    print("1 -- Vetor Reta-Reta")
    print("2 -- Vetor Reta-Plano")
    print("3 -- Vetor Plano-Plano")
    print("4 -- Sair")

    opcao = int(input("Digite a opcao desejada: "))
    if opcao == 1:
      vet_reta_reta()
    if opcao == 4:
      print("Saindo...")
      break
```
